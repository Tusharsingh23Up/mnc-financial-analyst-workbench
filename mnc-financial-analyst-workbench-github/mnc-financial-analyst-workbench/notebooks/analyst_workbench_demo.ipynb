```python
import pandas as pd
df = pd.read_csv('../data/financials_sample.csv')
df.head()
```

```python
df['fcf_margin'] = (df['cash_from_ops_usd_mm'] - df['capex_usd_mm']) / df['revenue_usd_mm']
df.groupby('ticker')['fcf_margin'].mean().sort_values(ascending=False)
```
